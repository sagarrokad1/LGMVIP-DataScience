### **Problem Statement:**

*   We need to read the image in RGB format and the convert it to a grayscale image. This will turn to an image into a classic black and white photo.

*   Then the next thing to do is invert the grayscale iimage also called as negative image, this will be our inverted grayscale image. Inversion can be used to enhance the details.

*   Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image.

*   Since the images are just arrays, we can easily do this programatically using the divide function from the cv2 library in Python.

### **Importing libraries**

```python
from google.colab.patches import cv2_imshow
import cv2
```

### **Load the Image**

```python
img = cv2.imread('/content/download.jpg', 1)

cv2_imshow(img)
```

###**Converting Image to Gray Scale**

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2_imshow(img_gray)
```

###**Invert the Image**

```python
img_invert = cv2.bitwise_not(img_gray)

cv2_imshow(img_invert)
```

### **Smoothening the image**

```python
img_smoothing = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)

cv2_imshow(img_smoothing)
```

### **Final sketch**

```python
def dodgeV2(x, y):

    return cv2.divide(x, 255 - y, scale=256)
```

```python
final_img = dodgeV2(img_gray, img_smoothing)

cv2_imshow(final_img)
```
